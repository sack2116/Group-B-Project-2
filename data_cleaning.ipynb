```python
from scipy.io import arff as arff_scipy
import pandas as pd

# Load the .arff file
data, meta = arff_scipy.loadarff('resources/dataset_44_spambase.arff')

# Convert to a Pandas DataFrame
df = pd.DataFrame(data)

# If there are byte strings (common in .arff files), convert them to regular strings
for column in df.select_dtypes([object]):
    df[column] = df[column].str.decode('utf-8')

# Display the DataFrame
display(df.head())
```
